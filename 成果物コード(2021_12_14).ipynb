```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
import statsmodels.api as sm
```

```python
df=df.set_index('都道府県')
```

```python
scaler.fit(np.array(df))
```

```python
std_df=scaler.transform(np.array(df))
```

```python
std_df=pd.DataFrame(std_df)
```

```python
std_df.columns=['人口','人口密度','平均年収','平均気温','飼育率']
```

```python
std_df
```

```python
s_x= std_df[['人口','人口密度','平均年収','平均気温']]
s_y= std_df['飼育率']
```

```python
model=sm.OLS(s_y,sm.add_constant(s_x))
```

```python
results=model.fit()
print(results.summary())
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
x1=std_df['人口']
x2=std_df['人口密度']
x3=std_df['平均年収']
x4=std_df['平均気温']
y=std_df['飼育率']
```

```python
fig=plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x1,y,'o')
plt.title('population vs pets')

plt.subplot(2,2,2)
plt.plot(x2,y,'o')
plt.title('density vs pets')

plt.subplot(2,2,3)
plt.plot(x3,y,'o')
plt.title('income vs pets')

plt.subplot(2,2,4)
plt.plot(x4,y,'o')
plt.title('temperature vs pets')

plt.tight_layout()
```
